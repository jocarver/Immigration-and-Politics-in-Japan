```python
import pandas as pd
```

```python
df = pd.read_csv("../../data/raw/Japan yearly population 2012-2024.csv", skiprows = 1)
```

```python
df.head()
```

```python
new_cols = []
for column in df.columns:
    new_cols.append(column.replace('-to', '').replace('-ken', '').replace('-fu', ''))
df.columns = new_cols
```

```python
df.rename(columns = {'Time': 'year', 'Gumma':'Gunma'}, inplace=True)
```

```python
df['year'] = pd.to_datetime(df['year'], format = '%Y')
```

```python
df.set_index('year', inplace=True)
```

```python
df.to_csv("../../data/clean/immigration_data/japan_pop_2012-2024.csv")
```
